```python
import opti
import pandas as pd
import numpy as np
from doe.design import find_local_max_ipopt
from doe.utils import ProblemProvider

pd.set_option('display.max_columns', 25)
pd.set_option('display.max_rows', 1000)
```

```python
problem = opti.Problem(
    inputs = [opti.Continuous(f"x{i+1}", [0,1]) for i in range(20)],
    outputs = [opti.Continuous("y")],
    constraints= [opti.LinearEquality(names=[f"x{i+1}" for i in range(20)], rhs=1)],
)

problem_provider = ProblemProvider(problem=problem)

res = find_local_max_ipopt(problem_provider=problem_provider, model_type="fully-quadratic", ipopt_options={"maxiter":200, "disp":100})
```

```python
np.round(res[0],2)
```
