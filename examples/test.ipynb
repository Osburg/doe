```python
import numpy as np
import doe
import pandas as pd
import opti
```

# Linear Models, no linear inequalities
## n_inputs = 20, mixture constraint

```python
problem = opti.Problem(
    inputs = [opti.Continuous(f"x{i+1}", [0,1]) for i in range(20)],
    outputs = [opti.Continuous("y")],
    constraints= [opti.LinearEquality(names=[f"x{i+1}" for i in range(20)], rhs=1)],
)

res = doe.find_local_max_ipopt(problem=problem, model_type="linear")
np.round(res,2)
```

# Linear Models, linear inequalities
# n_inputs = 20, first 5 inputs <= 0.5

```python
problem = opti.Problem(
    inputs = [opti.Continuous(f"x{i+1}", [0,1]) for i in range(20)],
    outputs = [opti.Continuous("y")],
    constraints= [
        opti.LinearEquality(names=[f"x{i+1}" for i in range(20)], rhs=1),
        opti.LinearInequality(names=["x1"], rhs=0.5),
        opti.LinearInequality(names=["x2"], rhs=0.5),
        opti.LinearInequality(names=["x3"], rhs=0.5),
        opti.LinearInequality(names=["x4"], rhs=0.5),
        opti.LinearInequality(names=["x5"], rhs=0.5),
    ],
)

res = doe.find_local_max_ipopt(problem=problem, model_type="linear")
np.round(res,2)
```
